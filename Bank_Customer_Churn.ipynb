```python
df['Exited_label'] = df['Exited'].replace({1: 'Churned', 0: 'Retained'})
```

```python
fig, ax = plt.subplots()

ax.hist(df['Exited_label'])
ax.set_title('Bank Customer Retained and Churned')
```

```python
data = df['Exited_label'].value_counts()
plt.pie(data, labels= [f'{data.index[0], data.values[0]}', f'{data.index[1], data.values[1]}'], autopct= '%1.2f%%')
plt.title('Bank Customer Retained and Churned')
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

ax[0].violinplot(df['CreditScore'], showmeans=True, showmedians=True)
ax[0].set_title('Credit Score')

ax[1].violinplot(df['EstimatedSalary'], showmeans=True, showmedians=True)
ax[1].set_title('EstimatedSalary')
```

```python
param = ['Age', 'Balance', 'CreditScore', 'EstimatedSalary', 'Point Earned']

fig, ax = plt.subplots(ncols=5, figsize=(15,4))

for i, col in enumerate(param):
  sns.histplot(df, x=col, ax=ax[i], kde=True)
  ax[i].set_title(f'Count by {col}')
```

```python
avg_data = df.groupby('Exited')[param].mean().reset_index()

fig, ax = plt.subplots(ncols=5, figsize=(16,4))

for i, col in enumerate(param):
  sns.barplot(data = avg_data, x='Exited', y=col, ax=ax[i], hue='Exited')
  ax[i].set_title(f'Average {col}')
  ax[i].set_xlabel('Churn')
  ax[i].set_ylabel('Average')
  
plt.tight_layout()
```

```python

param2 = ['Geography', 'Gender', 'HasCrCard', 'IsActiveMember', 'Card Type', 'Complain']

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(12,6))

for i, col in enumerate(param2):
  row = i // 3
  col_idx = i % 3
  ax[row, col_idx].pie(df[col].value_counts(), labels=df[col].unique(), autopct= '%1.2f%%')
  ax[row, col_idx].set_title(f'{col}')
```

```python
num_cols = 3
num_rows = (len(param2) + num_cols - 1) // num_cols 

fig, ax = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(12,6))

ax = ax.flatten()

for i, col in enumerate(param2):
  sns.countplot(data=df, x=col, ax = ax[i], hue='Exited')
  ax[i].set_title(f'Churn by {col}')
  ax[i].set_xlabel(col)
  ax[i].set_ylabel('Count')
  ax[i].tick_params(axis='x', rotation=0)

plt.tight_layout() 
```

```python
param3 = ['Tenure', 'NumOfProducts', 'Card Type']

fig, ax = plt.subplots(ncols=3, figsize=(10,4))

for i, col in enumerate(param3):
  sns.countplot(df, x=col, ax=ax[i])
  ax[i].set_title(f'{col}')
```

```python
correlation_matrix = df.corr(numeric_only=True)

fig, ax = plt.subplots(figsize=(10,8))

sns.heatmap(correlation_matrix, annot=True, ax=ax)
ax.set_title('Correlation Matrix')
plt.show()
```
